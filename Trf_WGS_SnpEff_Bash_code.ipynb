```bash
# Get annotation files and put into the snpeff data area
cd Zea_mays_B73_NAM.5.0
wget https://download.maizegdb.org/Zm-B73-REFERENCE-NAM-5.0/Zm-B73-REFERENCE-NAM-5.0_Zm00001eb.1.gff3.gz
mv Zm-B73-REFERENCE-NAM-5.0_Zm00001eb.1.gff3.gz genes.gff.gz

# Get the genome reference sequence file
cd /Users/aricha14/Documents/Tools/snpEff/data/genomes
wget https://download.maizegdb.org/Zm-B73-REFERENCE-NAM-5.0/Zm-B73-REFERENCE-NAM-5.0.fa.gz
mv Zm-B73-REFERENCE-NAM-5.0.fa.gz Zea_mays_B73_NAM.5.0.fa.gz
```

```bash
# Need to rename the fasta file to sequences.fa
cd /Users/aricha14/Documents/Tools/snpEff/data/Zea_mays_B73_NAM.5.0
mv Zea_mays_B73_NAM.5.0.fa.gz sequences.fa.gz
```

```bash
#Build the database 
cd /Users/aricha14/Documents/Tools/snpEff
java -jar snpEff.jar build -gff3 -v Zea_mays_B73_NAM.5.0
```

```bash
# Check to see if the correct genome is added 
java -jar /Users/aricha14/Documents/Tools/snpEff/snpEff.jar databases | grep -i Zea
```

```bash
# Now I have built the snpeff database I can run snpeff on the vcf files 
#Snpeff code 
cd /Users/aricha14/Documents/Research_Projects/TRF_Project/WGS/WGS_Alignment_NAM5_May2022
mkdir snpeff
java -Xmx4g -jar /Users/aricha14/Documents/Tools/snpEff/snpEff.jar Zea_mays_B73_NAM.5.0 /Users/aricha14/Documents/Research_Projects/TRF_Project/WGS/WGS_Alignment_NAM5_MAy2022/TrfM6_WGS_NAM5_mpileup_Dec2022.vcf > /Users/aricha14/Documents/Research_Projects/TRF_Project/WGS/WGS_Alignment_NAM5_May2022/snpeff/TrfM6_WGS_NAM5_mpileup_Dec2022.ann.vcf

```

```bash
#Check the output file of SnpEff to see the annotations

#Check the output file to make sure that it looks ok

awk '{if($1=="#CHROM") print $0}' ./snpeff/TrfM6_WGS_NAM5_mpileup_Dec2022.ann.vcf

head -10 ./snpeff/TrfM6_WGS_NAM5_mpileup_Dec2022.ann.vcf 
```

```bash
awk '{if($1=="#CHROM") print $0}' ./snpeff/TrfM6_WGS_NAM5_mpileup_Dec2022.ann.vcf  

awk '{if($1=="chr10") print $0}' ./snpeff/TrfM6_WGS_NAM5_mpileup_Dec2022.ann.vcf   |  head -10
```

```bash
 #Filter to just the variants on chromosome 10 
java -jar /Users/aricha14/Documents/Tools/snpEff/SnpSift.jar filter "(CHROM == 'chr10')" ./snpeff/TrfM6_WGS_NAM5_mpileup_Dec2022.ann.vcf    > ./snpeff/TrfM6_WGS_NAM5_mpileup_Dec2022_snpeff_chr10.ann.vcf 

```

```bash
# Then filter for the intreval region 13.5Mb to 14.0Mb on chr1 
java -jar /Users/aricha14/Documents/Tools/snpEff/SnpSift.jar filter "(CHROM == 'chr10') && (POS >= 13496844) && (POS < 14026688)" ./snpeff/TrfM6_WGS_NAM5_mpileup_Dec2022.ann.vcf  > ./snpeff/TrfM6_WGS_NAM5_mpileup_Dec2022_snpeff_chr10_AD2toAD3.ann.vcf

awk '{if($1=="#CHROM") print $0}' ./snpeff/TrfM6_WGS_NAM5_mpileup_Dec2022_snpeff_chr10_AD2toAD3.ann.vcf

awk '{if($1=="chr10") print $0}' ./snpeff/TrfM6_WGS_NAM5_mpileup_Dec2022_snpeff_chr10_AD2toAD3.ann.vcf  |  head -10
```

```bash
#Use SNPSift to filter for high impact variants, in the AD2 to AD3 interval

java -jar /Users/aricha14/Documents/Tools/snpEff/SnpSift.jar filter "ANN[*].IMPACT has 'HIGH' | ANN[*].IMPACT has 'MODERATE'" ./snpeff/TrfM6_WGS_NAM5_mpileup_Dec2022_snpeff_chr10_AD2toAD3.ann.vcf  > ./snpeff/TrfM6_WGS_NAM5_mpileup_Dec2022_snpeff_chr10_AD2toAD3_HIGHorMOD.vcf

awk '{if($1=="#CHROM") print $0}' ./snpeff/TrfM6_WGS_NAM5_mpileup_Dec2022_snpeff_chr10_AD2toAD3_HIGHorMOD.vcf

awk '{if($1=="chr10") print $0}' ./snpeff/TrfM6_WGS_NAM5_mpileup_Dec2022_snpeff_chr10_AD2toAD3_HIGHorMOD.vcf  |  head -10
```

```bash
# Repeat the process with the N pool so that we can compare which ones are present/ absent 
java -Xmx4g -jar /Users/aricha14/Documents/Tools/snpEff/snpEff.jar Zea_mays_B73_NAM.5.0 /Volumes/Elements/Sequence_Data/WGS_data/Trf_mutant_WGS_2022/TrfN5_WGS_NAM5_mpileup_Dec2022.vcf > /Users/aricha14/Documents/Research_Projects/TRF_Project/WGS/WGS_Alignment_NAM5_May2022/snpeff/TrfN5_WGS_NAM5_mpileup_Dec2022.ann.vcf

 #Filter to just the variants on chromosome 10 
java -jar /Users/aricha14/Documents/Tools/snpEff/SnpSift.jar filter "(CHROM == 'chr10')" ./snpeff/TrfN5_WGS_NAM5_mpileup_Dec2022.ann.vcf    > ./snpeff/TrfN5_WGS_NAM5_mpileup_Dec2022_snpeff_chr10.ann.vcf 

# Then filter for the interval region 110Mb to 165Mb on chr1 
java -jar /Users/aricha14/Documents/Tools/snpEff/SnpSift.jar filter "(CHROM == 'chr10') && (POS >= 13496844) && (POS < 14026688)" ./snpeff/TrfN5_WGS_NAM5_mpileup_Dec2022.ann.vcf  > ./snpeff/TrfN5_WGS_NAM5_mpileup_Dec2022_snpeff_chr10_AD2toAD3.ann.vcf

#Use SNPSift to filter for high impact variants, in the AD2 to AD3 interval

java -jar /Users/aricha14/Documents/Tools/snpEff/SnpSift.jar filter "ANN[*].IMPACT has 'HIGH' | ANN[*].IMPACT has 'MODERATE'" ./snpeff/TrfN5_WGS_NAM5_mpileup_Dec2022_snpeff_chr10_AD2toAD3.ann.vcf  > ./snpeff/TrfN5_WGS_NAM5_mpileup_Dec2022_snpeff_chr10_AD2toAD3_HIGHorMOD.vcf

```

```bash
#As it is RNAseq data and we want to account for splice junctions we use HiSAT2 to align the trimmed reads to B73v3 genome
find ./trimmed_reads -name "*_R1.fq.gz" | while read f; do
    STEM=$(basename "${f}" _R1.fq.gz)
    hisat2 --min-intronlen 20 --max-intronlen 20000 \
    -p 4 -x ./maize_genome/ZmaysAGPv3.30.hisat2\
    -1 ./trimmed_reads/"${STEM}"_R1.fq.gz \
    -2 ./trimmed_reads/"${STEM}"_R2.fq.gz \
    | samtools view -bS - \
    > ./aligned_reads/"${STEM}".bam \
    2> ./HiSAT2_report/"${STEM}"_summary.txt
done
```
